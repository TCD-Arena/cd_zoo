```python
from os import listdir
from os.path import isfile, join
import pandas as pd
import numpy as np
import pickle
import networkx as nx
import sys
sys.path.append("..")
from tools.tools import graph_to_label_tensor
```

```python
# Small example how to transfer from one format into the other. 
# For non overlapping samples, this format should NOT be chosen. 
```

```python
mypath = "split_format/"
onlyfiles = sorted([mypath + f for f in listdir(mypath) if isfile(join(mypath, f))])
```

```python
data = [pd.read_csv(x, index_col=0) for x in sorted([y for y in onlyfiles if "data" in y])]
data = pd.concat(data, axis=1)
data.columns = np.arange(len(data.columns))
```

```python
labels = [pd.read_csv(x, index_col=0) for x in sorted([y for y in onlyfiles if "label" in y])]
```

```python
labels = [pd.read_csv(x, index_col=0) for x in sorted([y for y in onlyfiles if "label" in y])]
a = np.arange(50)
for x in range(10):
    renaming =  a[5*x:5*x+5]
    labels[x].columns = renaming
    labels[x].index = renaming
```
